```python
# paths of other files
# path = ""
path = "../../../../data/flowResults/V5"   
```

```python
# loading all the files as pandas dataframes
# all_dfs = {<file name>: df}
all_dfs = {}

for filename in all_files:
    print(f'filename: {filename}')
    df = pd.read_csv(os.path.join(path, filename))
    print(type(df[' ipProto']))
    all_dfs[filename] = df
```

```python
# columns
'''
    Index(['srcMac', ' dstMac', ' ethType', ' srcIp', ' dstIp', ' ipProto',
        ' srcPort', ' dstPort', ' protocol', ' allMatches',
        ' numPayloadOnePackets', ' numPayloadTwoPackets'],
        dtype='object')
'''
ipProto = 'UDP'

if ipProto == 'TCP':
    ipProtoNum = 6
elif ipProto == 'UDP':
    ipProtoNum = 17

stdPort = '123'
protoName = 'ntp'
for filename, df in all_dfs.items():
    # getting the flows for this MUD flow
    # print(df[' ipProto'])
    # MUD_df = df.loc[df[' ipProto'].str.contains('6') & ((df[' srcPort']=='443') | (df[' dstPort']=='443'))]  #
    MUD_df_1 = df.loc[(df[' ipProto'] == ipProtoNum) & ((df[' srcPort']==stdPort) | (df[' dstPort']==stdPort)) & (df[' protocol'].str.contains(protoName))]  #
    MUD_df_2 = df.loc[(df[' protocol'].str.contains(protoName))]  #
    MUD_df_3 = df  #
    print(filename)
    print(f'{ipProto}/{stdPort} {protoName}:#######################')
    print(MUD_df_1.groupby(' protocol')[' protocol'].count().reset_index(name='No. of flows'))
    print(f'Total {protoName}:#######################')
    print(MUD_df_2.groupby(' protocol')[' protocol'].count().reset_index(name='No. of flows'))
    print(f'Total flows:#######################')
    print(f'\t\t     {len(MUD_df_3)}')
    print()
    print('\n')
    # break
```
