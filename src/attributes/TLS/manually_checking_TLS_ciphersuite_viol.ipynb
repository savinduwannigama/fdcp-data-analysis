## Client-hello-cipher-suites

```python
file_path = '../../../data/attributes/V3/request_attributes/client-hello-cipher-suites.csv'
final_output_file = '../../../results/attributes/TLS/client-hello-cipher-suites.csv'  
```

```python
'''
    dict01 = {<device_name>: <list of cipher suites: 2D list>}
'''
dict01 = {}

with open(file_path, 'r') as csvfile:
    reader = csv.reader(csvfile)

    prev_device = 'erorr: first line in the csv has no device'
    for row in reader:
        print('device:', row[0], end='  |  ')
        print('vals:', row[1], end='  |  \n')

        if row[0] == '':
            if row[1] != '':
                dict01[prev_device].append(row[1].split(','))  # appending because there already exists a list of cipher suites for the device
                print('empty device')
            else:
                continue
                print('both empty')
        else:
            print('both NOT empty')
            dict01[row[0]] = [row[1].split(',')]

        prev_device = row[0]
# print(dict01)
```

```python
# print(dict01)
for key, val in dict01.items():
    print(f'key: {key}  |  val: {val}')
```

```python
'''
    viol_dict = {
        <device>: {
            <cipher index>: {
                viols: <list of violating ciphers> OR "no_viol",
                frac: <"# violating ciphers" + / + "# of total ciphers in cipher suite">
            }
        }
    }
'''
viol_dict = {}
for device, listOfCS in dict01.items():
    viol_dict[device] = {}
    for i, cs in enumerate(listOfCS):  # cs is a list of ciphers
        no_weak = True
        no_insecure = True

        viol_dict[device][i] = {}

        weak_ciphers = []
        insecure_ciphers = []
        for c in cs:
            if len(list(c)) > 4:
                print(f'device: {device}, CS index: {i}, C: {c}')
            if c in weak_ciphersuites_codes:
                weak_ciphers.append(c)
                no_weak = False
            if c in insecure_ciphersuites_codes:
                insecure_ciphers.append(c)
                no_insecure = False
                # break
        if no_weak:
            viol_dict[device][i]['weak'] = 'no_viol'
            viol_dict[device][i]['weak/total'] = "0/" + str(len(cs))
        else:
            viol_dict[device][i]['weak'] = weak_ciphers
            viol_dict[device][i]['weak/total'] = str(len(weak_ciphers)) + "/" + str(len(cs))

        if no_insecure:
            viol_dict[device][i]['insecure'] = 'no_viol'
            viol_dict[device][i]['insecure/total'] = "0/" + str(len(cs))
        else:
            viol_dict[device][i]['insecure'] = insecure_ciphers
            viol_dict[device][i]['insecure/total'] = str(len(insecure_ciphers)) + "/" + str(len(cs))

        
        # viol_dict[device][i]['violating_ciphers'] = violating_ciphers
        # viol_dict[device][i]['violating'] = str(len(violating_ciphers)) + "/" + str(len(cs))

            
```

```python
for k, v in viol_dict.items():
    print(f'key: {k}  |  val: {v}')
```

```python
df_to_save = pd.DataFrame(viol_dict)
# Write the DataFrame to the CSV file
df_to_save.to_csv(final_output_file, index=True, sep='|')
```
