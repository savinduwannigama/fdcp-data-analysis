```python
import requests


```

```python
# Base URL of the Ciphersuite.info API
base_url = "https://ciphersuite.info/api"

# Endpoint for getting secure ciphersuites
secure_endpoint = "/cs/security/weak"

# Full URL for the secure ciphersuites endpoint
url = base_url + secure_endpoint

# Make the GET request
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Get the JSON data from the response
    weak_ciphersuites = response.json()

    # Now you have the list of secure ciphersuites
    print(f'weak_ciphersuites: {weak_ciphersuites}')
    print(len(weak_ciphersuites['ciphersuites']))
else:
    print("Failed to retrieve secure ciphersuites. Status code:", response.status_code)


secure_endpoint = "/cs/security/insecure"

# Full URL for the secure ciphersuites endpoint
url = base_url + secure_endpoint

# Make the GET request
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Get the JSON data from the response
    insecure_ciphersuites = response.json()

    # Now you have the list of secure ciphersuites
    print('insecure_ciphersuites: {insecure_ciphersuites}')
    print(len(insecure_ciphersuites['ciphersuites']))
else:
    print("Failed to retrieve secure ciphersuites. Status code:", response.status_code)
```

```python
# processing weak cipher suites
weak_ciphersuites_codes = []
for el in weak_ciphersuites['ciphersuites']:  # iterates a list
    for csname, csdict in el.items():  # will iterate only once
        hex_code = (csdict['hex_byte_1'].split('x')[1] + csdict['hex_byte_2'].split('x')[1]).lower()
        weak_ciphersuites_codes.append(hex_code)

        # debug
        print(f'csname: {csname}, hex_code: {hex_code}')


print(f'weak_ciphersuites_codes: {len(weak_ciphersuites_codes)}')
```

```python
# processing insecure cipher suites
insecure_ciphersuites_codes = []
for el in insecure_ciphersuites['ciphersuites']:  # iterates a list
    for csname, csdict in el.items():  # will iterate only once
        hex_code = (csdict['hex_byte_1'].split('x')[1] + csdict['hex_byte_2'].split('x')[1]).lower()
        insecure_ciphersuites_codes.append(hex_code)

        # debug
        print(f'csname: {csname}, hex_code: {hex_code}')


print(f'insecure_ciphersuites_codes: {len(insecure_ciphersuites_codes)}')
```
